```python
df=pd.read_csv(r'/Users/binodrai/Desktop/1718991964134-5938-HR-Data.csv')
```

```python
# IQR METHOD ::Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Monthly Income'].quantile(0.25)
Q3 = df['Monthly Income'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
###Outlier IQR score 
Outlier_zscore
Outlier_iqr = df[(df['Monthly Income'] >=upper_bound)|(df['Monthly Income'] <=lower_bound)]
Outlier_iqr
```

```python
###tRIMMING IQR score

# Using IQR for trimming, assuming we already have lower_bound and upper_bound calculated
trimmed_df = df[(df['Monthly Income'] <=upper_bound) & (df['Monthly Income'] >=lower_bound)]

trimmed_df.shape
```

```python
##Capping using IQR
df['Monthly Income'] = np.where(
    df['Monthly Income']>upper_bound,
    upper_bound,
    np.where(
        df['Monthly Income']<lower_bound,
        lower_bound,
        df['Monthly Income']
    )
)
```
